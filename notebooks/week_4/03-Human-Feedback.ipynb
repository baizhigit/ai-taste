### Import Dependencies

```python
from langsmith import Client
```

### Attach Feedback to a specific trace in LangSmith

019e075c-cc32-7670-8ac0-6721f04a3e42

```python
client = Client()
```

```python
RUN_ID = "019e075c-cc32-7670-8ac0-6721f04a3e42"
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comment",
    value="This is a long form free text feedback left by user",
    feedback_source_type="api"
)
```

```python
client.create_feedback(
    run_id=RUN_ID,
    key="comments_with_score",
    value="This is a long form free text feedback left by user with score",
    score=1,
    feedback_source_type="api"
)
```
